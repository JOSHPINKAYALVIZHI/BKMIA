```python
import pandas as pd
import matplotlib as mat
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/content/Bookjk.csv")
df
```

```python
mat.pyplot.scatter(df['year'],df['index'],color='purple')
mat.pyplot.xlabel('Year')
mat.pyplot.ylabel('Index')
mat.pyplot.title('AI in future Engineering jobs')
```
